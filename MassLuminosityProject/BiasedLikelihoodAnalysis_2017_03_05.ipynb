```python
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
import numpy as np

data = np.loadtxt('/Users/user/Code/Scratch/lines.txt', delimiter=',')
```

```python
a1 = data[:10]
a2 = data[10:20]
a4 = data[20:30]
S = data[30:]
```

```python
plt.rcParams['figure.figsize'] = (8, 8)

plt.subplot(221)
plt.title('alpha1 likelihood')
plt.plot(a1[:,0], a1[:,1])
plt.xlabel('alpha1')
plt.ylabel('log-likelihood')

plt.subplot(222)
plt.title('alpha2 likelihood')
plt.plot(a2[:,0], a2[:,1])
plt.xlabel('alpha2')
plt.ylabel('log-likelihood')

plt.subplot(223)
plt.title('alpha4 likelihood')
plt.plot(a4[:,0], a4[:,1])
plt.xlabel('alpha4')
plt.ylabel('log-likelihood')

plt.subplot(224)
plt.title('S likelihood')
plt.plot(S[:,0], S[:,1])
plt.xlabel('S')
plt.ylabel('log-likelihood')
plt.tight_layout()
```

## Vectorized Numpy Operations Avoid Excessive Bounds-Checking

```python
import numpy as np

a = np.random.rand(10**7)
b = np.random.rand(10**7)
c = np.zeros(10**7)

def example1():
    c=a+b
    
def example2():
    for i in xrange(len(a)):
        c[i] = a[i] + b[i]

def example3():
    tmp = 0
    for i in xrange(len(a)):
        tmp += 1
```

```python
%timeit -r 1 example1()
```

```python
%timeit -r 1 example2()
```

```python
%timeit -r 1 example3()
```

## Finite Precision Breaks Commutivity

$\pm\Big(\sum_{i=0}^{p-1} a_i \beta^{-i} \Big)\beta^e$

```python
from math import pi
x = (pi + 10**17) - 10**17
print np.cos(x)
print np.cos(pi)
```
